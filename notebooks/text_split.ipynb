```python
import duckdb
import pandas as pd

non_processed_pdfs_sql="""
SELECT *  
FROM papers
INNER JOIN papers_raw_md USING (id) 
WHERE papers.extract_done=TRUE AND papers.transform_done=FALSE;
"""

with duckdb.connect('../data.duckdb')  as  con:
    nonprocessed_pdfs=con.execute(non_processed_pdfs_sql).df()

```

```python
nonprocessed_pdfs.iloc[2].id.item()
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter


text_splitter = RecursiveCharacterTextSplitter(
    # Set a really small chunk size, just to show.
    separators=["\n\n", "\n", "."],
    chunk_size=1000,
    chunk_overlap=300,
    length_function=len,
    is_separator_regex=True,
)

text=nonprocessed_pdfs.iloc[2].raw_md.decode("utf-8")
text=text[:text.find("References")]
# text

texts = text_splitter.create_documents([text])

sep="\n#@#@#@#@#@#@#@#@#@#@#@#@#@#@#@#@#@#@#@#@#@#@#@#@#@#@#@#@#@#@#@#@#@#@#@#@#@#@#@#@#@#@#@#@#@#@#@#@#@\n"
print(len(texts))
[print(t.page_content,sep) for t in texts];
```
